```python
import pandas as pd     
import numpy as np           
import matplotlib.pyplot as plt
```

```python

Trans1 = pd.DataFrame({"Technique": ["Mu", "CWEh", "CWEhp", "Sim"],      "Value": [0.55, 0.55, 0.45, 0.64]})
Trans2 = pd.DataFrame({"Technique": ["Mu", "CWEh", "InMu", "InCWEh"],    "Value": [0.54, 0.52, 0.57, 0.56]})
GNN =    pd.DataFrame({"Technique": ["CVE", "CWE", "CVE+CWE", "Sample"], "Value": [0.88, 0.86, 0.89, 0.86]})




# Combine the data
GNN['Type'] = 'GNN'
Trans1['Type'] = 'Trans1'
Trans2['Type'] = 'Trans2'


combined_data = pd.concat([GNN,Trans1, Trans2])

plt.figure(figsize=(4.2, 4.2))
colors = {'Trans1': 'indigo', 'Trans2':  'brown', 'GNN': 'black'} # {'Trans1': 'blue', 'Trans2': 'orange', 'GNN': 'green'}


for key, group in combined_data.groupby('Type'):
    plt.bar(group['Technique'], group['Value'], color=colors[key], label=key)

plt.ylabel('MCC')
plt.title('MCC Values by incorporation technique') #  in and enhanced tranformers and GNN models
plt.xticks(rotation=45)
plt.legend(title='Technique')


plt.tight_layout()
plt.savefig('mcc_values.pdf')

plt.show()
```
